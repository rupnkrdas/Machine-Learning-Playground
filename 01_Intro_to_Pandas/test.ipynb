# Subsetting Data

```python
df.shape # Display the shape of the dataframe
```

```python
# Subsetting Columns
# Syntax: df[['column1', 'column2', 'column3']]
# Note: The double brackets are required
# Note: The order of the columns in the list is the order they will appear in the dataframe
df = df[['title', 'description', 'publishTime',
         'duration_seconds', 'viewCount', 'likeCount',
         'commentCount']] # Subset the dataframe to only include the columns listed
```

```python
df.head()
```

```python
# Subsetting using loc
# loc is used to subset rows and columns by label
# Syntax: df.loc[rows, columns]
df_subset1 = df.loc[df['viewCount'] > 1_000_000] # Subset the dataframe to only include rows where viewCount is greater than 1 million
df_subset1.head()
```

```python
# Subsetting using query
# query is used to subset rows and columns by label
# Syntax: df.query('column > value')
df_subset2 = df.query('viewCount > 1000000') # Subset the dataframe to only include rows where viewCount is greater than 1 million
df_subset2.head()
```

```python
df = df.loc[~df['likeCount'].isna()] # Subset the dataframe to only include rows where likeCount is not null
df.head()
```

# Casting dtypes

```python
# Casting Data Types
# Syntax: df['column'] = df['column'].astype('type')
df['viewCount'] = df['viewCount'].astype('int') # Convert the viewCount column to an integer
df['likeCount'] = df['likeCount'].astype('int') # Convert the likeCount column to an integer
```

```python
df['publishTime'] = pd.to_datetime(df['publishTime']) # Convert the publishTime column to a datetime
```

```python
df['likeCount'] = pd.to_numeric(df['likeCount'].astype('str')) 
```

# Creating new column

```python
# Creating New Columns
# Syntax: df['new_column'] = df['column1'] + df['column2']
df['like_to_view_ratio'] = df['likeCount'] / df['viewCount'] # Create a new column called like_to_view_ratio that is the likeCount divided by the viewCount
```

# Adding new Row

```python
# Creating New Rows
df_to_append = df.tail(1)
```

```python
df_concat = pd.concat([df, df_to_append])
```

```python
df.plot(kind='scatter', x='viewCount', y='likeCount', title='View vs Like Count')   # Plot a scatter plot of viewCount vs likeCount
                                                                                    # kind='scatter' specifies that we want a scatter plot
                                                                                    # x='viewCount' specifies that we want the viewCount column to be the x-axis
                                                                                    # y='likeCount' specifies that we want the likeCount column to be the y-axis
                                                                                    # title='View vs Like Count' specifies the title of the plot
```
